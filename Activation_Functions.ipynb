## Activation Functions

Activation functions are mathematical equations that determine the output of a neural network. The function is attached to each neuron in the network, and determines whether it should be activated (“fired”) or not, based on whether each neuron’s input is relevant for the model’s prediction. Activation functions also help normalize the output of each neuron to a range between 1 and 0 or between -1 and 1. [[4](biblio.txt)]

```python
# Import matplotlib, numpy, and keras
import matplotlib.pyplot as plt 
import numpy as np 
import keras

%matplotlib inline
```

### Step Function

```python
# if 0 > x:
#      return 0
# elif x >= 0:
#      return 1

x = np.linspace(-10, 10, 100) 
z = np.piecewise(x, [0 > x, x >= 0], [0, 1])
  
plt.plot(x, z) 
plt.xlabel("x") 
plt.ylabel("Step(x)") 
  
plt.show()
```

### Sigmoid Function

```python
x = np.linspace(-10, 10, 100) 
z = 1/(1 + np.exp(-x)) 
  
plt.plot(x, z) 
plt.xlabel("x") 
plt.ylabel("Sigmoid(x)") 
  
plt.show()
```

### Tanh Function

```python
x = np.linspace(-10, 10, 100) 
z = np.tanh(x)
  
plt.plot(x, z) 
plt.xlabel("x") 
plt.ylabel("Tanh(x)") 
  
plt.show()
```

### Rectified Linear Unit (ReLU) Function

```python
x = np.linspace(-10, 10, 100) 
z = np.maximum(0,x)

plt.plot(x, z) 
plt.xlabel("x") 
plt.ylabel("ReLU(x)") 
  
plt.savefig("figures/relu")
```

## Let's compare a couple of these to activation functions in Keras

```python
x = np.linspace(-10, 10, 100) 
z = keras.activations.tanh(x)
  
plt.plot(x, z) 
plt.xlabel("x") 
plt.ylabel("Tanh(x)") 
  
plt.show() 
```

```python
x = np.linspace(-10, 10, 100) 
z = keras.activations.relu(x)
  
plt.plot(x, z) 
plt.xlabel("x") 
plt.ylabel("ReLU(x)") 
  
plt.show() 
```
