This is a test notebook for downloading images from ImageNet

```python
import requests
```

```python
# Grab Images
not_hotdog = {
    "chicken-sandwhich": "http://image-net.org/api/text/imagenet.synset.geturls?wnid=n07696728",
    "pizza": "http://image-net.org/api/text/imagenet.synset.geturls?wnid=n07873807",
    "fries": "http://image-net.org/api/text/imagenet.synset.geturls?wnid=n07711232",
    "taco": "http://image-net.org/api/text/imagenet.synset.geturls?wnid=n07880751",
    "bacon": "http://image-net.org/api/text/imagenet.synset.geturls?wnid=n07862244",
    "chicken-kiev": "http://image-net.org/api/text/imagenet.synset.geturls?wnid=n07864756",
    "baked-potato": "http://image-net.org/api/text/imagenet.synset.geturls?wnid=n07710952",
    "burrito": "http://image-net.org/api/text/imagenet.synset.geturls?wnid=n07880968",
    "sushi": "http://image-net.org/api/text/imagenet.synset.geturls?wnid=n07879450",
    "dessert": "http://image-net.org/api/text/imagenet.synset.geturls?wnid=n07609840",
    "vegetables": "http://image-net.org/api/text/imagenet.synset.geturls?wnid=n07708398",
    "fruit": "http://image-net.org/api/text/imagenet.synset.geturls?wnid=n13134947",
    
}
```

```python
for food_name, url in not_hotdog.items():
    for i, image_url in enumerate(requests.get(url).iter_lines()):
        if image_url:
            try:
                image_request = requests.get(image_url, timeout=3)
            except requests.exceptions.SSLError:
                print("SSL Error - moving on")
            except requests.exceptions.Timeout:
                print("Timeout - moving on")
            except requests.exceptions.ConnectionError:
                print("Connection Error - moving on")
        if image_request.status_code == requests.codes.ok:
            with open(f"data/images/not_hotdog/{i}-{food_name}.jpg", "wb") as f:
                f.write(image_request.content)
                print(f"Saved {image_url}")
```
